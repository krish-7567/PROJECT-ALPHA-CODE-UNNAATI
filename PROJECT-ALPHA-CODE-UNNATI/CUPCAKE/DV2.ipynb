```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("updated_data.csv")

# Create a heatmap (example: brand vs. price ranges)

```

```python
#Bar Chart: Smartphone Brand Popularity
#What it conveys: This graph shows the number of smartphones listed for each brand.
#It helps identify the most dominant and least represented brands in the dataset.
#Use case: Useful for understanding brand presence or market dominance on Flipkart.

# Pivot table for the heatmap
pivot_table = data.pivot_table(index="brand", columns="battery_capacity", values="ratings", aggfunc="mean")

# Plot heatmap
sns.heatmap(pivot_table, annot=True, cmap="coolwarm", fmt=".1f", linewidths=.5)
plt.title("Heatmap: Ratings by Brand and Battery Capacity")
plt.show()

```

```python
# Histogram: Price Distribution
# What it conveys: This graph displays the distribution of smartphone prices. 
# Peaks in the histogram can reveal popular price ranges, while gaps indicate underrepresented price points.
# Use case: Helpful for pricing strategies or analyzing buyer preferences for budget, mid-range, or premium phones.

# Count brands
brand_counts = data["brand"].value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
brand_counts.plot(kind="bar", color="skyblue")
plt.title("Smartphone Brand Popularity")
plt.xlabel("Brand")
plt.ylabel("Number of Smartphones")
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter Plot: Ratings vs. Price
# What it conveys: This plot shows how smartphone ratings vary with price. Patterns may reveal whether higher-priced phones consistently receive better ratings or if ratings are independent of cost.
# Use case: Provides insight into customer satisfaction across different price segments.
# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(data["discounted_price"], bins=20, color="orange", edgecolor="black")
plt.title("Price Distribution of Smartphones")
plt.xlabel("Price (INR)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Line Plot: Battery Capacity vs. Price
# What it conveys: This graph demonstrates how battery capacity affects average smartphone prices. A positive correlation may suggest that higher battery capacities are valued more.
# Use case: Useful for understanding which features (like battery capacity) drive higher pricing.
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data["discounted_price"], data["ratings"], color="green", alpha=0.6)
plt.title("Ratings vs Price")
plt.xlabel("Price (INR)")
plt.ylabel("Ratings")
plt.show()
```

```python
# Pie Chart: Battery Type Distribution
# What it conveys: This chart breaks down the share of each type of battery in the dataset. It shows the prevalence of different battery types (e.g., Li-ion, Li-polymer).
# Use case: Helps manufacturers or retailers gauge which battery types are more common or popular.
# Line plot
avg_prices = data.groupby("battery_capacity")["discounted_price"].mean()

plt.figure(figsize=(10, 6))
plt.plot(avg_prices.index, avg_prices.values, color="blue", marker="o")
plt.title("Battery Capacity vs Price")
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Average Price (INR)")
plt.grid()
plt.show()
```

```python
# Box Plot: Price by Brand
# What it conveys: This graph visualizes the price range (minimum, median, maximum) for each brand. It highlights pricing strategies and reveals outliers (e.g., exceptionally expensive or cheap models).
# Use case: Helps compare how brands position themselves in terms of pricing.
# Pie chart
battery_counts = data["battery_type"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(battery_counts, labels=battery_counts.index, autopct="%1.1f%%", colors=["skyblue", "orange", "lightgreen"])
plt.title("Battery Type Distribution")
plt.show()
```

```python
# Count Plot: Rear Camera Configurations
# What it conveys: This plot counts how often each unique rear camera setup (e.g., dual, triple cameras) appears in the dataset.
# It shows the popularity of different configurations.
# Use case: Useful for identifying trends in camera setups and consumer preferences.

plt.figure(figsize=(12, 6))
sns.boxplot(x="brand", y="discounted_price", data=data, palette="Set3")
plt.title("Price Distribution by Brand")
plt.xlabel("Brand")
plt.ylabel("Price (INR)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap: Correlation Matrix
# What it conveys: The heatmap highlights correlations between numerical columns like discounted_price, ratings, 
#  rating_count, and battery_capacity. Strong correlations may indicate how different factors influence each other.
# Use case: Provides a quick overview of relationships in the data to guide deeper analysis.
# Correlation heatmap
correlation_matrix = data[["discounted_price", "ratings", "rating_count", "battery_capacity"]].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Line Plot: Display Size vs. Price
# What it conveys: This graph shows how display size influences smartphone prices. 
#   Larger displays may correlate with higher prices.
# Use case: Helps understand how specific features (like screen size) contribute to pricing decisions.
# Line plot
avg_price_display = data.groupby("display_size")["discounted_price"].mean()

plt.figure(figsize=(10, 6))
plt.plot(avg_price_display.index, avg_price_display.values, marker="o", color="purple")
plt.title("Display Size vs Price")
plt.xlabel("Display Size (inches)")
plt.ylabel("Average Price (INR)")
plt.grid()
plt.show()
```
